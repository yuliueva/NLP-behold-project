# Feature Cleaning

```python
clean_data = data_cleaned.copy()
clean_data = clean_data.drop(columns = 'product_color_id')
clean_data = clean_data.drop_duplicates()
```

## Lemmatization

```python
lemmatizer = WordNetLemmatizer()

for i in range(len(clean_data)):
    review = clean_data.iloc[i,1].split()
    new_words = []
    for word in review:
        new_words.append(lemmatizer.lemmatize(word))
    cleaned_review = " ".join(new_words)
    clean_data.iloc[i,1] = cleaned_review
```

```python
lemmatizer = WordNetLemmatizer()

for i in range(len(clean_data)):
    review = clean_data.iloc[i,2].split()
    new_words = []
    for word in review:
        new_words.append(lemmatizer.lemmatize(word))
    cleaned_review = " ".join(new_words)
    clean_data.iloc[i,2] = cleaned_review
```

## Merge Data

```python
merge = clean_data.groupby(['product_id','attribute_name'])['attribute_value'].apply(lambda x: ','.join(x)).reset_index()
```

```python
merge2 = merge.drop(columns = 'product_id')
merge2 = merge2.drop_duplicates()
attr_name = merge2.groupby(['attribute_name'])['attribute_value'].apply(lambda x: ','.join(x)).reset_index()
value_name = attr_name.attribute_value.apply(lambda x: x.split(',')).apply(lambda x: set(x))
```

```python
final = merge.pivot(index='product_id', columns='attribute_name', values='attribute_value').reset_index().rename_axis(None, axis=1)
```
